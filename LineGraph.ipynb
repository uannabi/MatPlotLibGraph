```python
plt.figure(figsize=(10,5))
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#Solid Blue line will be plotted using the argument "b-"
plt.figure(figsize=(10,5))
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y,'-b')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# Solid red line will be plotted using the argument "r-"
plt.figure(figsize=(10,5))
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y,'-r')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# Plot Green dots using the argument "go"
plt.figure(figsize=(10,5))
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y,'go')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```
